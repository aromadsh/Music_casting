```python

# sunflower_url = "./test_img.png"

def voice (x):
    y, sr = librosa.load(x)
    S = librosa.feature.melspectrogram(y, sr=sr)
    S_DB = librosa.amplitude_to_db(S, ref=np.max)
    img_array = librosa.display.specshow(S_DB, sr=sr, hop_length=512)
    plt.savefig("t.png")
    img = Image.open("./t.png")
    img = np.array(img)
    print(img.shape)
    class_names = ['J-POP', 'OST', 'POP', 'R&B_Soul', '댄스', '랩_힙합',\
     '록_메탈', '발라드', '성인가요', '인디음악', '일렉트로니카', '포크_블루스', '포크_블루스_컨트리']
    img = keras.preprocessing.image.load_img("./t.png", target_size=(653, 979))
    img_array = keras.preprocessing.image.img_to_array(img)
    img_array = tf.expand_dims(img_array, 0) 
    img_array = np.array(img_array)
    print(img_array.shape)
    VGGmodel2 = load_model('./data/model/220927b8e10.hdf5')
    predictions = VGGmodel2.predict(img_array)
    plt.imshow(img)
    score = tf.nn.softmax(predictions[0])
    
    return class_names[np.argmax(score)]
```

```python
voice("./승민.wav")
```
